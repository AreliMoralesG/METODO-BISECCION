```
#TAREA 3 METODO DE BISECCION# 

import matplotlib.pyplot as plt
import numpy as np

def f(x):
    y= x**2 + 6*x**3 + 8
    return y

a=0 #punto inicial 
b=1 #punto final 
tol=1e-10 #tolerancia 
n0=50
i=1
fa=f(a)
fb=f(b)

while i<=n0:

    p=a+(b-a)/2

    fp=f(p)

    if fp == 0 or (b-a)/2<tol:
        print("La solucion es ",p,"despues de ",i,"iteraciones")
        break

    i=i+1
    if fa*fp>0:
        a=p
        fa=fp

    else:
        b=p
        fb=fp

if i>=n0:
    print("El metodo no tiene buena aproximacion")
  
x=np.linspace(1, 6, 300)
plt.plot(x, f(x))
plt.grid()
plt.show()
```
